```python
import matplotlib.pyplot as plt
import numpy as np
import math
from mpl_toolkits.mplot3d import Axes3D

n = 10
x = [i/n for i in range(0,n+1)]
y = [i/n for i in range(0,n+1)]
z = [[math.sin(math.pi*(i/n)**1)*math.sin(math.pi*(j/n)**1) for i in range(0,n+1)] for j in range(0,n+1)]
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(x, y, z, 50, cmap='inferno')
plt.show()

f = open("V_1.txt")
r = f.read().split('\n')
f.close()
# print(r)

a = [[0. for i in range(0,n+1)] for j in range(0,n+1)]
for k in range (0, 10+1):
    for i in range (0, n+1):
        for j in range (0, n+1):
            a[i][j] = float(r[i*(n+1)+j + k*(n+1)**2])
    print(a[5][5] , (z[5][5]*k/n))
    if (k!=0):
        fig = plt.figure()
        ax = plt.axes(projection='3d')
        ax.contour3D(x, y, a, 50, cmap='inferno')
        plt.show()
```
